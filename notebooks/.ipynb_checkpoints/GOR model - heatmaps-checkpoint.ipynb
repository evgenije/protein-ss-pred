```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename_model = '../models/model.npz'

index = ['A', 'R', 'N', 'D', 'C', 'Q', 'E', 'G', 'H', 'I', 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V']
columns = range(-8,9)
```

```python
model = np.load(filename_model)['model'].item()
```

```python
def get_heatmap(data, index, columns):
    df = pd.DataFrame(data, index=index, columns=columns)
    plt.figure(figsize=(10, 10))
    sns.heatmap(df)
    plt.yticks(rotation=0)
```

```python
# Plotting the heatmap for helix in the GOR model
get_heatmap(model['H'].transpose(), index, columns)
```

```python
# Plotting the heatmap for strand in the GOR model
get_heatmap(model['E'].transpose(), index, columns)
```

```python
# Plotting the heatmap for coil in the GOR model
get_heatmap(model['-'].transpose(), index, columns)
```
