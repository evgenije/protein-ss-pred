```python
import matplotlib.pyplot as plt
```

```python
id_list_filename_training = '../data/training/list.txt'
dssp_dir_training = '../data/training/dssp'
id_list_filename_blindTest = '../data/blindTest/list.txt'
dssp_dir_blindTest = '../data/blindTest/dssp'

structures_names = ['Helix', 'Strand', 'Coil']
```

```python
def get_ss_counts_file(dssp_filename, ss={}):
    with open(dssp_filename) as dssp_f:
        dssp_f.readline()
        dssp = dssp_f.readline().rstrip().replace('C', '-')
        for char in dssp:
            ss[char] = ss.get(char,0) + 1
    return ss
```

```python
def get_ss_counts_dir(id_list_filename, dssp_dir):
    ss = {}
    with open(id_list_filename) as id_list:
        for line in id_list:
            ss = get_ss_counts_file(dssp_dir+'/'+line.rstrip()+'.dssp', ss)
    return ss
```

```python
def get_pie_chart(counts, names):
    values = list(counts)
    total = sum(values)
    for i in range(0, len(values)):
        values[i] = round(values[i]/float(total)*100, 2)
    labels = []
    for i in range(0, len(values)):
        labels.append(names[i]+' '+str(values[i])+'%')
    plt.figure(figsize=(10, 10))
    plt.pie(values, labels=labels, startangle=60)
    plt.axis('equal')
```

```python
# Plotting the pie chart for secondary structure counts in the training set
ss = get_ss_counts_dir(id_list_filename_training, dssp_dir_training)
structures = [ss.get('H', 0), ss.get('E', 0), ss.get('-', 0)]
get_pie_chart(structures, structures_names)
```

```python
# Plotting the pie chart for secondary structure counts in the blind test set
ss = get_ss_counts_dir(id_list_filename_blindTest, dssp_dir_blindTest)
structures = [ss.get('H', 0), ss.get('E', 0), ss.get('-', 0)]
get_pie_chart(structures, structures_names)
```
